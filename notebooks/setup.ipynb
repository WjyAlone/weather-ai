```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
np.random.seed(5)
x = np.linspace(-1, 1, 100).reshape(100,1)
y = 3*np.power(x, 2) + 2 + 0.2*np.random.rand(x.size).reshape(100,1)
plt.scatter(x, y)
plt.show()
```

```python

w1 = np.random.rand(1)
b1 = np.random.rand(1)
lr = 0.001
for epoch in range(1000):
    y_pred = np.power(x, 2)*w1+b1
    loss = 0.5*(y_pred-y)**2
    loss = loss.sum()
    gred_w = np.sum((y_pred-y)*np.power(x, 2))
    gred_b = np.sum(y_pred-y)
    w1 -= lr * gred_w
    b1 -= lr*gred_b
plt.plot(x, y_pred, 'r-', label='predict')
plt.scatter(x, y, color='blue', marker='o', label='true')
plt.xlim(-1, 1)
plt.ylim(2, 6)
plt.legend()
plt.show()
print(w1, b1)
```
